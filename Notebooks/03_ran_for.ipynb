### **Summary: Random Forest as a combination of regression trees**

A regression tree consists in subdividing the input space into partitions using a set of decision rules.

At each node $i$ of the tree, the feature dimension $d_i$ of the input vector $x$ is compared to a threshold value $t_i$.

**For instance:** 
Consider an input vector $x = (x_1, x_2)$. The first node asks if $x_1$ is less than some threshold $t_1$. If yes, we then ask if $x_2$ is less than some other threshold $t_2$. If yes, we enter the bottom-left leaf node. This corresponds to the region of space defined by:

$$R_1 = \{x : x_1 \le t_1, x_2 \le t_2\}$$

Each region can be associated with a predicted output, and we can then partition the entire input space into $J$ disjoint regions.

Formally, a regression tree can be defined by:

$$f(x) = \sum_{j=1}^{J} w_j I(x \in R_j)$$

where $R_j$ is the region specified by the $j$-th leaf node, $I(\cdot)$ is the indicator function, and $w_j$ is the predicted output for that node. The predicted output $w_j$ is simply the average of the responses of the training samples falling in the region $R_j$:

$$w_j = \frac{\sum_{n=1}^{N} y_n I(x_n \in R_j)}{\sum_{n=1}^{N} I(x_n \in R_j)}$$
